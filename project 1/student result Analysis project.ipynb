```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv.zip")
print(df.head())
```

# drop unnamed column


```python
df=df.drop("Unnamed: 0" , axis = 1)
print(df.head())
```

# change wklystudyhours column

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-","5-oct")
df.head()
```

# gender distribution



```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distributition")
plt.show()
```

# from the above chart we have analyed that the number of female in the data is more than the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot = True)
plt.title("Relation b/w Parent's Education and student Score")
plt.show()
```

```python
#from the above chart we havre concluded that the education of the parents have a good empect score
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot = True)
plt.show()
```

# distribution og ethnic groups

```python
groupA=df.loc[(df["EthnicGroup"] == "group A")].count()
groupB=df.loc[(df["EthnicGroup"] == "group B")].count()
groupC=df.loc[(df["EthnicGroup"] == "group C")].count()
groupD=df.loc[(df["EthnicGroup"] == "group D")].count()
groupE=df.loc[(df["EthnicGroup"] == "group E")].count()

l=["groupA","groupB","groupC","groupD","groupE"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of group")
plt.show()
```
